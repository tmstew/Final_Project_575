```python
import numpy as np
import pandas as pd
```

```python
ref = pd.read_table('23andme_v5_hg19_ref.txt', header = None)
```

```python
ref.columns = ['Chromosome', 'Position','Variant','Allele']
```

```python
ref
```

```python
ann = pd.read_csv('annotations/var_drug_ann.tsv', sep = '\t', error_bad_lines = False)
#Any better options?
```

```python
ann
```

```python
combine  = pd.merge(ref, ann)
```

```python
combine
```

```python
combine_1 = combine.loc[: ,['Variant','Gene','Chemical', 'Literature ID', 'Phenotype Category','Significance','Notes','Sentence','Alleles','Allele' ]]
```

```python
combine_1
```

```python
combine_1.columns = ['dbSNP_ID','GENE_SYMBOL','DRUG_NAME','PMID','PHENOTYPE_CATEGORY','SIGNIFICANCE','NOTES','SENTENCE', 'ALLELE_PharmGKB','ALLELE_23andme']
```

```python
combine_1
```

```python
filtered = combine_1.loc[(combine_1.PHENOTYPE_CATEGORY == 'efficacy') & (combine_1.SIGNIFICANCE =='yes')]
```

```python
filtered
```

```python
df_final  = filtered.loc[: ,['dbSNP_ID','GENE_SYMBOL','DRUG_NAME','NOTES','SENTENCE', 'ALLELE_PharmGKB','ALLELE_23andme']]
```

```python
df_final
```

```python
# make drug_name cleaner (in the same format)
for elem in df_final.index:
    DRUG = df_final.loc[elem, 'DRUG_NAME']
    df_final.loc[elem, 'DRUG_NAME'] = DRUG.replace('"','')
```

```python
#replace NaN to blank string
df_final = df_final.fillna('')
```

```python
df_final
```

```python
df_final.to_csv('23andme_PharmGKB_map.tsv', sep="\t")
```

```python
#extra credit
df_extra = df_final.loc[:,['dbSNP_ID','GENE_SYMBOL','DRUG_NAME']]
```

```python
df_extra
```

```python
group = df_extra.groupby(['GENE_SYMBOL','DRUG_NAME'])['dbSNP_ID'].apply(';'.join)
```

```python
df_group = pd.DataFrame(group)
```
