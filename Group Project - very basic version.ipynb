```python
import numpy as np
import pandas as pd
```

```python
data23me_names = ['CHR','POS','dbSNP_ID','ALLELE']
data23me = pd.read_csv('/Users/wangshuchun/Desktop/575FinalProject/23andme_v5_hg19_ref.txt', sep = '\t', names = data23me_names)
```

```python
data_var_drug = pd.read_csv('/Users/wangshuchun/Desktop/575FinalProject/var_drug_ann.tsv', sep = '\t', usecols = [i for i in range(12)])
```

```python
new_names = ['dbSNP_ID','GENE_SYMBOL','DRUG_NAME','PMID','PHENOTYPE_CATEGORY','SIGNIFICANCE','NOTES','SENTENCE','ALLELE_PharmGKB','ALLELE_23andme']
whole_merged_df = pd.DataFrame(columns = new_names)
for db_SNP in data23me['dbSNP_ID']:
    data_index = data_var_drug[data_var_drug.Variant == db_SNP].index
    if len(data_index) != 0:
        for i in range(len(data_index)):
            allele23me = data23me.ALLELE[data23me.dbSNP_ID == db_SNP]
            Data_series = data_var_drug.loc[data_index[i],'Variant':'Alleles'].append(allele23me).drop('StudyParameters')
            names_dict = dict(zip(Data_series.index, new_names))
            single_df = Data_series.to_frame().T.rename(columns = names_dict)
            whole_merged_df = whole_merged_df.append(single_df)
print(whole_merged_df)
```

```python
whole_merged_df = whole_merged_df[(whole_merged_df['SIGNIFICANCE'] == 'yes') & (whole_merged_df['PHENOTYPE_CATEGORY'] == 'efficacy')]
whole_merged_df
```

```python
whole_merged_df.to_csv('23andme_PharmGKB_map.tsv', sep = '\t', columns = ['dbSNP_ID', 'GENE_SYMBOL', 'DRUG_NAME', 'NOTES', 'SENTENCE', 'ALLELE_PharmGKB', 'ALLELE_23andme'])
```
